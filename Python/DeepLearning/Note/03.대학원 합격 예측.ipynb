```python
import pandas as pd
data = pd.read_csv('../Data/gpascore.csv')
data.head()
```

- admit : 합격여부
- gre : 영어성적
- gpa : 학점
- rank : 지원한 대학원 수준

### Preprocessing(전처리)

```python
# 결측치
data.isnull().sum()
```

```python
# 결측치 데이터 삭제
data.dropna(inplace=True)
data.isnull().sum()
```

```python
# rank의 종류
sorted(data['rank'].unique())
```

```python
# 영어 점수의 최소값과 최대값
print('최대값 :', data['gre'].max())
print('최소값 :', data['gre'].min())
```

```python
# 합격여부 데이터 빈도수 확인
print('합격개수 :', data[data['admit'] == 1].count())
print('불합격개수 :', data[data['admit'] == 0].count())
```

```python
data[data['admit'] == 1].count() / data[data['admit'] == 0].count()
```

### Train Data 만들기

```python
x = data[['gre','gpa','rank']]
x.head()
```

```python
y = data['admit']
y.head()
```

## Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(256, activation='tanh', input_shape=(3,)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import numpy as np
```

```python
model.fit(np.array(x), np.array(y), epochs=500, verbose=1)
```

```python
model.evaluate(np.array(x), np.array(y))
```

```python
# 예측하기
# gre : 700, 학점 : 3.7, Rank 4학교?
model.predict([[700,3.7,4]])
```
