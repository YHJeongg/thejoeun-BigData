```python
station_files = glob('../../Data/oil/*.xls')
station_files
```

```python
tmp_raw = []
for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
station_raw = pd.concat(tmp_raw)
station_raw.head(50)
```

```python
len(station_raw)
```

```python
station_raw.info()
```

```python
# 필요 항목 발췌
stations = pd.DataFrame({
    'Oil_store' : station_raw['상호'],
    '주소' : station_raw['주소'],
    '가격' : station_raw['휘발유'],
    '셀프' : station_raw['셀프여부'],
    '상표' : station_raw['상표']
})

stations.head()
```

```python
stations.info()
```

```python
# 주유소별 자치구 항목 추가
stations['구'] = [region.split()[1] for region in stations['주소']]
stations
```

```python
# 구 컬럼 확인
stations['구'].unique()
```

```python
# 여러개 파일을 합하여 만든 데이터셋이기 때문에 index가 중복되어 있다.
stations.reset_index(inplace=True, drop=True)
stations
```

```python
top_station = stations.sort_values(by='가격', ascending=False).head(10)
low_station = stations.sort_values(by='가격', ascending=False).tail(10)
```

```python
import urllib.request

client_id = "-"
client_secret = "-"
index = 1
lng = []
lat = []
for starbucksAddress in top_station['주소']:
    try:
        encText = urllib.parse.quote(starbucksAddress)
        url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query=" + encText # json 결과
        request = urllib.request.Request(url)
        request.add_header("X-NCP-APIGW-API-KEY-ID",client_id)
        request.add_header("X-NCP-APIGW-API-KEY",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        if(rescode==200):
            response_body = response.read()
            entext = response_body.decode('utf-8')
        else:
            print("Error Code:" + rescode)

        import json    
        data1 = json.loads(entext)
        lng.append(data1['addresses'][0]['x'])
        lat.append(data1['addresses'][0]['y'])
        print(index, ': 주소 -', starbucksAddress, ', 경도 -', data1['addresses'][0]['x'], ', 위도 -', data1['addresses'][0]['y'])
        index += 1
    except:
        print(index, ': Error')
        pass

print('Done')
```

```python
top_station['위도'] = lat
top_station['경도'] = lng
top_station
```

```python
low_station['위도'] = lat
low_station['경도'] = lng
low_station
```

```python
seoul_map = folium.Map(
    location=[37.55, 126.98],
    tiles='Stamen Terrain',
    zoom_start=12
)

for lat, lng in zip(top_station.위도, top_station.경도):
    folium.Marker([lat, lng], icon=folium.Icon(icon = 'star', color = 'red')).add_to(seoul_map)
for lat, lng in zip(low_station.위도, low_station.경도):
    folium.Marker([lat, lng], icon=folium.Icon(icon = 'heart', color = 'blue')).add_to(seoul_map)

seoul_map
```
