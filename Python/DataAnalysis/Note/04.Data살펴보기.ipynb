### Pandas의 내장 그래프 도구

```python
# 남북한의 발전량 데이터셋 불러오기
df = pd.read_excel('../Data/남북한발전전력량.xlsx', engine='openpyxl')
df.head()
```

```python
# 남한, 북한 발전량 합계 데이터만 추출하여 df_ns로 만들기
df_ns = df[df['발전 전력별'] == '합계']
df_ns
```

```python
# df_ns의 행인덱스를 South와 North로 변경하기
df_ns.index = ['South', 'North']
```

```python
df_ns = df_ns.drop(['전력량 (억㎾h)','발전 전력별'], axis=1)
df_ns
```

```python
# 선 그래프
df_ns.T.plot()
```

```python
# 막대 그래프
df_ns.T.plot(kind='bar')
```

```python
tdf_ns = df_ns.T
```

```python
# 히스토그램
tdf_ns.plot(kind='hist')
```
