```python
import nltk
import pickle
from nltk.corpus import stopwords
import re
nltk.download('all')
```

```python
content = open('../Data/Independence.txt').read()
content
```

```python
# 불필요한 성별 없에기
cleaned_content = re.sub(r'[^\.\?\!\w\d\s]','',content)
cleaned_content
```

```python
# 대문자 소문자로 변환
cleaned_content = cleaned_content.lower()
cleaned_content
```

```python
# 각각의 워드를 토큰으로 쪼갠다.
word_tokens = nltk.word_tokenize(cleaned_content)
word_tokens
```

```python
# 품사분리
tokens_pos = nltk.pos_tag(word_tokens)
tokens_pos
```

```python
# 명사만 추출하기
NN_words = []
for word, pos in tokens_pos:
    if 'NN' in pos:
        NN_words.append(word)

print(NN_words)
```

```python
# 원형으로 복원
wlem = nltk.WordNetLemmatizer()
lemmatized_words = []
for word in NN_words:
    new_word = wlem.lemmatize(word)
    lemmatized_words.append(new_word)
print(lemmatized_words)
```

```python
# 불용어 제거
stopwords_list = stopwords.words('english')
unique_NN_words = set(lemmatized_words)
finally_NN_words = lemmatized_words

for word in unique_NN_words:
    if word in stopwords_list:
        while word in finally_NN_words:
            finally_NN_words.remove(word)

print(finally_NN_words)
```

```python
# 원하는 단어 제거
customized_stopwords = ['be', 'today', 'yesterday', "it's", "don't"]
unique_NN_words1 = set(lemmatized_words)

for word in unique_NN_words1:
    if word in customized_stopwords:
        while word in finally_NN_words:
            finally_NN_words.remove(word)

print(finally_NN_words)
```

```python
# 빈도 분석
from collections import Counter
c = Counter(finally_NN_words)
print(c)
```

```python
k = 20
print(c.most_common(k))
```

### WordCloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wordcloud
```

```python
noun_text = ''
for word in finally_NN_words:
    noun_text = noun_text + ' ' + word
wordcloud = WordCloud(
    max_font_size = 60,
    relative_scaling = .5
).generate(noun_text)

plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
