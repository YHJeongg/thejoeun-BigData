### 동일한 그림에 여러개의 그래프를 추가

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출
mask = (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul.drop(['전출지별'],axis=1,inplace=True)
df_seoul.rename({'전입지별' : '전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)
df_seoul
```

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구 데이터 값만 선택
df_3 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], :]
df_3
```

```python
# 컬럼이름 : 문자
col_years = list(map(str, range(1970,2018)))
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

ax.plot(
    df_3.columns,
    df_3.loc['충청남도'],
    marker='o',
    markerfacecolor='green',
    markersize=10,
    color='olive',
    linewidth=2,
    label='서울->충남'
)
ax.plot(
    df_3.columns,
    df_3.loc['경상북도'],
    marker='o',
    markerfacecolor='blue',
    markersize=10,
    color='skyblue',
    linewidth=2,
    label='서울->경북'
)
ax.plot(
    df_3.columns,
    df_3.loc['강원도'],
    marker='o',
    markerfacecolor='hotpink',
    markersize=10,
    color='pink',
    linewidth=2,
    label='서울->강원'
)

ax.legend(loc='best')
ax.set_title('서울->충남, 경북, 강원 인구 이동', size=20)
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)
ax.set_xticklabels(df_3.columns, rotation=90)

plt.show()
```

```python
def plotSeoul(inputAx, city, faceColor, colors):
    inputAx.plot(
        df_3.columns,
        df_3.loc[city],
        marker='o',
        markerfacecolor=faceColor,
        markersize=10,
        color=colors,
        linewidth=2,
        label='서울 -> '+city
    )
    inputAx.set_ylim(5000, 70000)
    inputAx.legend(loc='best')
    inputAx.set_title('서울->' + city + ' 인구 이동', size=12)
    inputAx.set_xticklabels(df_3.columns, rotation=45, size=8)
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

plotSeoul(ax1, '충청남도', 'green', 'olive')
plotSeoul(ax2, '경상북도', 'blue', 'skyblue')
plotSeoul(ax3, '강원도', 'hotpink', 'pink')
plotSeoul(ax4, '전라남도', 'yellow', 'orange')

plt.show()
```

### 2개의 축을 갖는 그래프 작성
- 남북한 발전량
- 막대그래프는 북한의 발전량
- 선 그래프는 북한 발전량의 전년 대비 증감률을 백분율로 표시

```python
df = pd.read_excel('../Data/남북한발전전력량.xlsx', convert_float=True)
df
```

```python
# 북한 전력량만 선택
df = df.loc[5:9]
df
```

```python
# 전력량 (억kWh) 컬럼 삭제
df.drop("전력량 (억㎾h)", axis=1, inplace=True)
df
```

```python
# 발전 전력별을 index로 변경
df.set_index('발전 전력별', inplace=True)
df
```

```python
# 기준점을 발전 전력별로 변경
df = df.T
df
```

```python
# 합계 컬럼을 총발전량으로 변경
df = df.rename(columns={'합계':'총발전량'})
df
```

```python
df['이전년도 총발전량'] = df['총발전량'].shift(1) # shift: 전(prev)행의 데이터를 선택
df
```

```python
# 증감율 컬럼 색성
df['증감율'] = (df['총발전량'] / df['이전년도 총발전량'] - 1) * 100
df
```

```python
plt.style.use('ggplot')

ax1 = df[['수력','화력']].plot(
    kind='bar',
    figsize=(20,10),
    stacked=True
)

ax2 = ax1.twinx()
ax2.plot(
    df.index,
    df.증감율,
    ls='--',
    marker='o',
    markersize=10,
    color='green',
    label='전년대비 증감율(%)'
)

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량', size=20)
ax2.set_ylabel('증감율', size=20)

plt.title('북한 전력 발전량(1990 ~ 2016)', fontsize=30)
ax1.legend(loc='upper left', fontsize=20)
ax2.legend(loc='best', fontsize=20)

plt.show()
```

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
df.head()
```

```python
# 자동차의 국가별 데이터 개수를 파이차트로 표현
# 국가별 Count를 계산하기 위해
df['count'] = 1
df.head()
```

```python
df_origin.index = ['USA','EU','JAPAN']
df_origin
```

```python
df_origin['count'].plot(
    kind='pie',
    figsize=(7, 5),
    autopct='%.2f%%',
    startangle=30,
    colors=['chocolate', 'bisque', 'cadetblue']
)

plt.title('Model Origin', size=20)
plt.axis('equal')
plt.legend(
    labels=df_origin.index,
    loc='upper right'
)

plt.show()
```
