```python
data = [
    [1.4,np.nan],
    [7.1,-4.5],
    [np.nan,np.nan],
    [0.75,-1.3]
]

df = pd.DataFrame(data,columns=['one','two'],index=['a','b','c','d'])
df
```

```python
# 열방향 합계 구하기
df.sum(axis='rows')
```

```python
# 행방향 합계 구하기
df.sum(axis='columns')
```

```python
# 열방향 평균 구하기
df.mean(axis='rows')
```

```python
# 행방향 평균 구하기
df.mean(axis='columns', skipna=False)
```

```python
# df one컬럼의 NaN : one의 평균값으로 대체
df.loc[df.one.isnull(), 'one'] = df.mean(axis='rows')['one']
df
```

```python
# df two컬럼의 NaN : two의 최소값으로 대체
df.fillna(value=df.min(axis='rows')['two'], inplace=True)
df
```

### DataFrame Merging(병함)

```python
df1 = pd.DataFrame({
    'Key' : list('bbacaab'),
    'data1' : range(7)
})

df1
```

```python
df2 = pd.DataFrame({
    'Key' : list('abd'),
    'data2' : range(3)
})

df2
```

```python
# default : Inner Join
pd.merge(df1, df2, on='Key')
```

```python
# 모두 보이기
pd.merge(df1, df2, on='Key', how='outer')
```

```python
# df1 기준으로 합치기
pd.merge(df1, df2, on='Key', how='left')
```

```python
# df2 기준으로 합치기
pd.merge(df1, df2, on='Key', how='right')
```

### Key Name이 다를 경우

```python
df1 = pd.DataFrame({
    'lkey' : list('bbacab'),
    'data1' : range(6)
})

df1
```

```python
df2 = pd.DataFrame({
    'rkey' : list('ababd'),
    'data1' : range(5)
})

df2
```

```python
pd.merge(df1, df2, left_on='lkey', right_on='rkey', how='inner')
```

### DataFrame의 Concatenation

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3,2),
    index=['a','b','c'],
    columns=['one','two']
)

df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2,2),
    index=['a','c'],
    columns=['three','four']
)

df2
```

```python
# df1과 df2 합치기
pd.concat([df1, df2], axis=1, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

### 데이터프레임 중복값 제거

```python
# 중복값 확인
df.duplicated()
```

```python
 # 중복값 제거
 df.drop_duplicates()
```

```python
# 새로운 열 추가
df['v1'] = np.arange(7)
df
```

```python
df['k1'].drop_duplicates()
```

```python
# k1 값들로 중복의 마지막 값 출력
df.drop_duplicates(['k1'], keep='last')
```

### Category

```python
df3 = pd.DataFrame({
    'id' : [1,2,3,4,5,6],
    'raw_grade' : ['a','b','b','a','a','e']
})

df3
```

```python
# category 자료형으로 변환하기
df3['grade'] = df3['raw_grade'].astype('category')
df3
```

```python
# category를 이용하여 자료 변형
df3.grade.cat.categories = ['very good', 'good', 'very bed']
df3
```

```python
df3.sort_values(by='grade')
```

```python
df3.grade.info()
```

### 데이터의 범위 정하기

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32]
bins=[18,25,35,60,100]
```

```python
cats = pd.cut(ages, bins)
cats
```

```python
# 범주 관련 Code 보기
cats.codes
```

```python
# 범주 관련 빈도수 보기
cats.value_counts()
```

```python
# category 이름 정하기
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
cat2 = pd.cut(ages, bins, labels=group_names)
cat2.value_counts()
```
