### 특성(Future) 컬럼 생성

```python
# FamilySize라는 컬럼을 생성 : SibSp + Parch + 1
train['FamilySize'] = train.SibSp + train.Parch + 1
train[['SibSp','Parch','FamilySize']].head()
```

```python
# Embarked에 따른 컬럼을 생성
# Q, S => UK, C => Australia
train['Nationality_AT'] = (train.Embarked == 'C')
train['Nationality_UK'] = (train.Embarked != 'C')
```

```python
train[['Embarked','Nationality_AT','Nationality_UK']].head()
```

```python
train['Nationality'] = ['England' if i else 'Austria' for i in train.Nationality_UK]
```

```python
train[['Embarked','Nationality_AT','Nationality_UK','Nationality']].head()
```

```python
train.loc[train['Embarked'] == 'C']
```

```python
# Fare에 대해 분류 컬럼을 만들려고 한다. 기준점 ?
train.Fare.describe()
```

```python
print('1사분위수 :', train.Fare.quantile(.25))
print('2사분위수 :', train.Fare.quantile(.5))
print('3사분위수 :', train.Fare.quantile(.75))
print('4사분위수 :', train.Fare.quantile(1))
```

```python
# 2분위수와 3분위수의 근사 정수치를 선택
train['LowFare'] = train['Fare'] < 15
train['MediumFare'] = (train['Fare'] >= 15) & (train['Fare'] < 30)
train['HighFare'] = train['Fare'] >= 30

train[['Fare','LowFare','MediumFare','HighFare']].head()
```

```python
train.loc[train['LowFare'] == True, 'FareType'] = 'Low'
train.loc[train['MediumFare'] == True, 'FareType'] = 'Medium'
train.loc[train['HighFare'] == True, 'FareType'] = 'High'

train[['Fare','FareType']].head(10)
```

```python
# Python의 함수를 이용하여 처리하기

def check_fare_type(fare):
    if fare < 15:
        return 'Low'
    elif fare >= 15 and fare < 30:
        return 'Medium'
    else:
        return 'High'
```

```python
train['FareType2'] = train['Fare'].apply(check_fare_type)
train[['Fare','FareType','FareType2']].head(10)
```

```python
# Age의 Nan을 Age의 평균값으로 대체하기 (단 소수점 첫째자리 까지 반올림 하기)
train.loc[train.Age.isnull(), 'Age'] = round(train.Age.mean())
train.Age.isnull().sum()
```
