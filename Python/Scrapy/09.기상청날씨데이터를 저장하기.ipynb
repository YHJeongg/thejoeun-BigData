```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")
html = req.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
for i in soup.find_all('location'):
    print(i.find('city').text)
```

```python
# 서울로 테스트
seoul = soup.find('location')
# tmef: 날짜, wf: 날씨, tmn: 최저기온, tmx: 최고기온
for i in seoul.find_all('data'):
    print(i.find('tmef').text)
    print(i.find('wf').text)
    print(i.find('tmn').text)
    print(i.find('tmx').text)
```

```python
# 도시별 날씨 데이터 구축
weather = {}

for i in soup.find_all('location'):
    weather[i.find('city').text] = []
    for j in i.find_all('data'):
        temp = []
        temp.append(j.find('tmef').text)
        temp.append(j.find('wf').text)
        temp.append(j.find('tmn').text)
        temp.append(j.find('tmx').text)
        weather[i.find('city').text].append(temp)

weather
```

```python
# csv file로 저장하기
# 현재 dictionary로 되어 있으나 List형태로 변환이 필요
for i in weather:
    for j in weather[i]:
        print(i, j[0], j[1], j[2], j[3])
    print('-' * 40)
```

```python
# List로 생성하기
tempList = []
for i in weather:
    for j in weather[i]:
        tempList.append((i, j[0], j[1], j[2], j[3]))
tempList
```

```python
import pandas as pd
```

```python
weatherDataFrame = pd.DataFrame(tempList, columns=['지역', '날짜', '날씨', '최저온도', '최고온도'])
weatherDataFrame.head()
```

```python
# File로 저장
weatherDataFrame.to_csv('../Data/weatherList.csv', index=False)
```

```python
 # sqlite 추가하기
 import sqlite3
```

```python
dbPath = '../Data/weatherList.sqlite'
conn = sqlite3.connect(dbPath)

curs = conn.cursor()
curs.executescript(
    """
    /* items 테이블이 이미 있다면 제거하기 */
    DROP TABLE IF EXISTS weatherList;

    /* 테이블 생성하기 */
    CREATE TABLE weatherList (
        seq INTEGER PRIMARY KEY AUTOINCREMENT,
        city TEXT,
        date DATETIME,
        weather TEXT,
        minTemp int,
        maxTemp int
    );
    """
)

conn.commit()
```

```python
sql = 'INSERT INTO weatherList (city, date, weather, minTemp, maxTemp) VALUES (?, ?, ?, ?, ?)'
for i in weatherDataFrame.itertuples():
    curs.execute(sql, (i[1],i[2],i[3],i[4],i[5]))
conn.commit()
```

```python
curs.execute('SELECT * FROM weatherList')
item_list = curs.fetchall()
for i in item_list:
    print(i)
```
