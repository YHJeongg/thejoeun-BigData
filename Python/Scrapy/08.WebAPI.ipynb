```python
api = 'http://api.openweathermap.org/data/2.5/weather?q={city}&APPID={key}'
```

### Naver API로 위도 경도 가져오기

```python
import urllib.request

client_id = "-"
client_secret = "-"


encText = urllib.parse.quote("울산광역시 동구 방어진순환도로 879 (서부동)")
url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query=" + encText # json 결과
request = urllib.request.Request(url)
request.add_header("X-NCP-APIGW-API-KEY-ID",client_id)
request.add_header("X-NCP-APIGW-API-KEY",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    entext = response_body.decode('utf-8')
else:
    print("Error Code:" + rescode)

import json    
data1 = json.loads(entext)
print(data1)
```

```python
import pandas as pd
```

```python
starbucks = pd.read_csv('../Data/Starbucks.csv')
starbucks['Address'] = starbucks['Address'].replace('"','')
starbucks['Address'][0]
```

```python
client_id = "-"
client_secret = "-"
index = 1
lng = []
lat = []
for starbucksAddress in starbucks['Address']:
    try:
        encText = urllib.parse.quote(starbucksAddress)
        url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query=" + encText # json 결과
        request = urllib.request.Request(url)
        request.add_header("X-NCP-APIGW-API-KEY-ID",client_id)
        request.add_header("X-NCP-APIGW-API-KEY",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        if(rescode==200):
            response_body = response.read()
            entext = response_body.decode('utf-8')
        else:
            print("Error Code:" + rescode)

        import json    
        data1 = json.loads(entext)
        lng.append(data1['addresses'][0]['x'])
        lat.append(data1['addresses'][0]['y'])
        print(index, ': 주소 -', starbucksAddress, ', 경도 -', data1['addresses'][0]['x'], ', 위도 -', data1['addresses'][0]['y'])
        index += 1
    except:
        print(index, ': Error')
        pass

print('Done')
```

```python
starbucks['Longitude'] = pd.Series(lng)
starbucks['Latitude'] = pd.Series(lat)
starbucks
```

```python
starbucks.to_csv("../Data/Starbucks_Address.csv", encoding='utf-8', index=False)
```
