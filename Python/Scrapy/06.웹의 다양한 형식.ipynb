### XML분석
기상청 전국 날씨 분류하기

```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
# 기상청 RSS 사이트 주소
url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
res = req.urlopen(url)
soup = BeautifulSoup(res, 'html.parser')
print(soup)
```

```python
test = {}
test['apple'] = []
test['apple'].append(1)
test['apple'].append(2)

test['banana'] = []
test['banana'].append(1)
test['banana'].append(2)

print(test)
```

```python
info = {}
for location in soup.find_all('location'):
    # print(location)
    name = location.find('city').text
    weather = location.find('wf').text
    if not(weather in info):
        info[weather] = []
    info[weather].append(name)

print(info)
```

```python
for i in info.keys():
    print('+', i)
    for j in info[i]:
        print('| -', j)
```

### JSON

```python
import json
```

```python
url = 'https://zeushahn.github.io/Test/student.json'
res = req.urlopen(url)

soup = BeautifulSoup(res, 'html.parser')
# soup을 string으로 변경해서 json형태로 변형하기
items = json.loads(str(soup))
print(items)
```

```python
for i in items['results']:
    for j in list(items['results'][0]):
        print(i[j], end=' ')
    print('\n')
```
