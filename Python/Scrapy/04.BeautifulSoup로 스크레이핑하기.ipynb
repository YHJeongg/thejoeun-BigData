### CSS 선택자 사용하기
- BeautifulSoup은 JavaScript Library인 JQuery처럼 CSS 선택자를 지정해서 원하는 요소를 추출 가능

```python
url = "https://zeushahn.github.io/Test/python/bs_exam04.html"

res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# 필요한 부분을 CSS Query로 추출하기 (#: id, .: class, >: 자식, 빈칸: 후손)
h1 = soup.select_one("div#meigen > h1").string
print("h1 = " + h1)

li_list = soup.select("div#meigen > ul.items > li")

for li in li_list:
    print("li = "+ li.string)
```

```python
url = "https://finance.naver.com/marketindex/"

res = req.urlopen(url)

soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
price = soup.select_one("#exchangeList > li.on > a.head.usd > div >span.value").string
print("usd/krw", price)
```

```python
for i in range(1, 5):
    price = soup.select_one(f"#exchangeList > li:nth-child({i}) > a.head > div > span.value").string
    country = soup.select_one(f"#exchangeList > li:nth-child({i}) > a.head > h3").string
    countryName = country.split()[0]
    print(f"{countryName} : {price}")
```

```python
prices = soup.select("#exchangeList span.value")
for price in prices:
    print(price.string)

countris = soup.select("#exchangeList h3")
for country in countris:
    print(country.string)
```

```python
#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a

url = "https://movie.daum.net/ranking/boxoffice/yearly"

res = req.urlopen(url)

soup = BeautifulSoup(res, "html.parser")
```

```python
movies = soup.select("#mainContent > div > div.box_boxoffice > ol  div > div.thumb_cont > strong > a")
index = 1
for movie in movies:
    print(index, ":", movie.string)
    index += 1
```

```python
imgs = soup.select("#mainContent > div > div.box_boxoffice > ol  div > div.thumb_item > div.poster_movie > img")
index = 1
for img in imgs:
    print(index, ":", img['alt'], "-", img['src'])
    index += 1
```

```python
# parse를 사용한 인코딩
from urllib import parse
```

```python
url = 'https://ko.wikisource.org/wiki/저자:윤동주'
url = parse.quote(url)
print(url)
```

```python
url = 'https://ko.wikisource.org/wiki/'
auth = '저자'
name = '윤동주'
url = url + parse.quote(auth) + ':' + parse.quote(name)
print(url)
res = req.urlopen(url)

soup = BeautifulSoup(res, "html.parser")
```

```python
#mw-content-text > div.mw-parser-output > ul:nth-child(6) > li > ul > li:nth-child(1) > a
novels = soup.select("#mw-content-text > div.mw-parser-output  ul  a")
index = '-'
for novel in novels:
    print(index, novel.string)
```

```python
url = "https://movie.daum.net/ranking/boxoffice/weekly"

res = req.urlopen(url)

soup = BeautifulSoup(res, "html.parser")
```

```python
movies = soup.select("#mainContent > div > div.box_boxoffice > ol  div > div.thumb_cont > strong > a")
index = 1
indexList = []
movieList = []
for movie in movies:
    indexList.append(index)
    movieList.append(movie.string)
    index += 1
print(indexList)
print(movieList)
```

```python
import pandas as pd
# countLists만으로 Dataframe 생성
df = pd.DataFrame(data = indexList, columns = ['Index'])
# df Dataframe에 컬럼 추가하기
df['Titles'] = pd.Series(movieList)
df
```

```python
df2 = pd.DataFrame(data=zip(indexList, movieList), columns=['Index', 'Title'])
df2
```

```python
# csv로 저장
df.to_csv("../Data/DaumList.csv", encoding='utf-8', index=False)
```
